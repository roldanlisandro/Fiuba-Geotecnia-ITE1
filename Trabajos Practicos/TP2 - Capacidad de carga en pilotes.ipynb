**Ayuda 1**:

Para estimar un G o E a una presión distinta a la de referencia use la siguiente expresión:

$E_{ur}(\sigma_3) = E_{ur}^{ref} \left( \frac{c \cdot cos \phi - \sigma_3 sin \phi}{c \cdot cos \phi + p_{ref} sin \phi}\right)^m$, con $m=0.5$


```python
def rigidez_sigma_3(E_ref,c,phi,sigma3,pref,m):
  """
  Inputs:
  E_ref o G_ref, en MPa (cualquier módulo)
  c, en kPa
  phi, en radianes!
  sigma3 o sigmah, en kPa
  pref, en kPa (adoptar pref=100)
  m (adoptar m=0.5)

  Outputs:
  E o G, en MPa (cualquier módulo), para una presión sigma3
  """
  
  E_sigma3 = E_ref * ((c*np.cos(phi) - sigma3*np.sin(phi))/((c*np.cos(phi) + pref*np.sin(phi))))**m
  return E_sigma3
```
