## Teoría

El ejericio consiste en realizar el cálculo de la resistencia por punta de un pilote utilizando la teoría de Brinch-Hansen. La teoría la vamos a repasar en clase, pero para este ejercicio solo necesitamos las fórmulas.

$Q_u=q_u A$ `carga última en la punta`

$q_u=c N_c s_c d_c + \sigma'_v N_q s_q d_q$ `tensión última en la punta`

$\sigma'_v = \gamma' z$ `tensión efectiva vertical`

$N_q = e^{\pi tan \phi}tan^2(\pi/4 + \phi/2)$ `coef. de capacidad de carga`

$N_c = (N_q-1) cot \phi $ `coef. de capacidad de carga`

$d_c = 1+0.4 tan^{-1}(z/D)$ `factor de profundidad`

$d_q = 1+2tan\phi (1-sin\phi)^2 tan^{-1}(z/D)$ `factor de forma`

$s_c = 1+N_q/N_c$ `factor de forma`

$s_q = 1+tan\phi $ `factor de forma`

$A$: area de la sección del pilote; $\gamma'$: peso efectivo del terreno; $\phi$: ángulo de fricción; $c$: cohesión
