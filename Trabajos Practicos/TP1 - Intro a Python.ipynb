## Teoría

El Factor de seguridad de un talud infinito (no sumergido) en condiciones drenadas viene dado por:

$FS = \frac{c+\gamma \cdot z \cdot cos^2(\beta) \cdot tan(\phi)}{\gamma \cdot z \cdot cos(\beta) \cdot sin(\beta)}$


Nota: cuando el suelo es friccionante ($c=0kPa$) el resultado es independiente de la profundidad y $FS = tan(\phi) / tan(\beta)$
